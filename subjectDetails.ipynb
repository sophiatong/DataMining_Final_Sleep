```python
details = pd.read_excel("files/SubjectDetails.xls")
details
```

```python
# plot BMI distribution 
# healthy BMI: 18.5 to 24.9

# Define two thresholds
threshold1 = 18.5
threshold2 = 24.9
# Create bins using pd.cut
bins = [-float('inf'), threshold1, threshold2, float('inf')]
# Create labels for the bins
labels = ['below', 'within', 'above']
# Add a new column 'Groups' based on binning
details['Groups'] = pd.cut(details['BMI'], bins=bins, labels=labels, right=False)

details
```

every single patient with apnea has a BMI higher than the healthy range!!!

```python
# quantiles of the BMI values
quantiles = [0.25, 0.5, 0.75]

# Calculate the quantiles of the 'Values' column
quantile_values = details['BMI'].quantile(quantiles)

# Display the calculated quantiles
print(f"Quantiles ({quantiles}):")
print(quantile_values)
```

```python
# Calculate the correlation between 'Epworth Sleepiness Score' and 'BMI'
correlation1 = details['Epworth Sleepiness Score'].corr(details['BMI'])
correlation2 = details['Epworth Sleepiness Score'].corr(details['Sleep Efficiency (%)'])
correlation3 = details['Sleep Efficiency (%)'].corr(details['BMI'])
correlation4 = details['Sleep Efficiency (%)'].corr(details['Age'])
# correlation5 = details['Sleep Efficiency (%)'].corr(details['Gender'])

# Display the correlation coefficient
print(f"Correlation between 'Epworth Sleepiness Score' and 'BMI': {correlation1}")
print(f"Correlation between 'Epworth Sleepiness Score' and 'Sleep Efficiency (%)': {correlation2}")
print(f"Correlation between 'Sleep Efficiency (%)' and 'BMI': {correlation3}")
print(f"Correlation between 'Sleep Efficiency (%)' and 'Age': {correlation4}")
# print(f"Correlation between 'Sleep Efficiency (%)' and 'Gender': {correlation5}")
```

```python
# read summarized respiratory events table
file_path = 'respevt_summary_df.csv'
respevt = pd.read_csv(file_path)
respevt
```

```python
respevt['Sum_HYP'] = respevt[['HYP-O_COUNT', 'HYP-C_COUNT', 'HYP-M_COUNT']].sum(axis=1, skipna=True)
respevt['Sum_APNEA'] = respevt[['APNEA-O_COUNT', 'APNEA-C_COUNT', 'APNEA-M_COUNT']].sum(axis=1, skipna=True)
respevt
```

```python
# correlation between apnea c duration and apnea m duration
correlation1 = respevt['APNEA-C_AVG_Duration'].corr(respevt['APNEA-M_AVG_Duration'])
# correlation between apnea C duration and apnea count
correlation2 = respevt['APNEA-C_AVG_Duration'].corr(respevt['Sum_APNEA'])
# correlation between apnea M duration and apnea count
correlation3 = respevt['APNEA-M_AVG_Duration'].corr(respevt['Sum_APNEA'])
#
correlation4 = respevt['Sum_HYP'].corr(respevt['Sum_APNEA'])

print(f"Correlation between apea C durtion and apnea M duration: {correlation1}")
print(f"Correlation between apea C durtion and apnea total count: {correlation2}")
print(f"Correlation between apea M durtion and apnea total count: {correlation3}")
print(f"Correlation between total hypnea count and apnea total count: {correlation4}")
```
